## Objective :
Build a classification model to predict the credibility of customers (default payment next month) based on demographic factors, credit data, history of payment, and bill statements.

## Dataset Information :
This dataset contains information on default payments, demographic factors, credit data, history of payment, and bill statements of credit card clients in Taiwan from April 2005 to September 2005.



## <font color='#3b82f6'>1. Importing Libraries</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

import warnings
warnings.filterwarnings('ignore')
```

## <font color='#3b82f6'>2. Load Data</font>

```python
df = pd.read_csv('Credit_Card_Default.csv')
df.head()
```

## <font color='#3b82f6'>3. Exploratory Data Analysis (EDA)</font>

```python
print("Dataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

## <font color='#3b82f6'>4. Target Variable Distribution</font>

```python
plt.figure(figsize=(6,4))
sns.countplot(x='default.payment.next.month', data=df, palette='viridis')
plt.title('Distribution of Default Payment')
plt.show()
```

## <font color='#3b82f6'>5. Data Preprocessing</font>

```python
# Mapping unknown categories
df['EDUCATION'] = df['EDUCATION'].replace([0, 5, 6], 4)
df['MARRIAGE'] = df['MARRIAGE'].replace(0, 3)

# Splitting Features and Target
X = df.drop(['ID', 'default.payment.next.month'], axis=1)
y = df['default.payment.next.month']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## <font color='#3b82f6'>6. Model Building and Overfitting Check</font>
To check for overfitting, we compare Training Accuracy vs Testing Accuracy.

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
    'SVM': SVC(random_state=42)
}

results = []
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    train_acc = accuracy_score(y_train, model.predict(X_train_scaled))
    test_acc = accuracy_score(y_test, model.predict(X_test_scaled))
    results.append({'Model': name, 'Train Accuracy': train_acc, 'Test Accuracy': test_acc})
    
    print(f"--- {name} ---")
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy:  {test_acc:.4f}\n")

results_df = pd.DataFrame(results)
```

## <font color='#3b82f6'>7. Model Building and Comparison</font>
Comparing the performance metrics of all models.

```python
print(results_df)
```

## <font color='#3b82f6'>8. Cross Validation</font>
Using 5-fold cross-validation to ensure model stability.

```python
cv_results = {}
for name, model in models.items():
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5)
    cv_results[name] = cv_scores
    print(f"{name} CV Mean Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
```

## <font color='#3b82f6'>9. Model Performance Comparison (Visualization)</font>

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Overfitting plot
results_df.set_index('Model').plot(kind='bar', ax=ax[0])
ax[0].set_title('Training vs Testing Accuracy')
ax[0].set_ylabel('Accuracy')

# CV plot
sns.boxplot(data=pd.DataFrame(cv_results), ax=ax[1])
ax[1].set_title('Cross-Validation Score Distribution')
ax[1].set_ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

## <font color='#3b82f6'>10. Conclusion</font>
Based on the analysis, we can determine the best performing model for credit card default prediction. Random Forest and SVM show competitive results, with Random Forest often providing a good balance between accuracy and stability.
